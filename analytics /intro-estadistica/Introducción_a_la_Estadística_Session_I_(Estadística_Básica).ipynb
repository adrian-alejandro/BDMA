muestral tiende a ser menos sesgada de forma sistemática.

De esta forma, para asegurarnos de que esto no ocurra, la muestra tiene que ser diversa (y representativa) y que los puntos/observaciones de la muestra sean independientes entre sí.

#### 7. Identifique si existe algún piso que podemos considerar “outlier” (para ello sólo tenga en cuenta las variables continuas). Elimínelo del data frame.

Para determinar los outliers de la muestra recurriremos al método LOF (Local Outlier Factor), el cual detecta potenciales outliers basándose en la densidad local de los puntos de la muestra.

```r
install.packages("DescTools")
```

```r
library(DescTools)

# Primero nos quedamos con las variables continuas:

cols <- c('Valor', 'Superf', 'Edat', 'ValSol')

data <- bcn_pisos[, cols]

# Calculamos los LOF scores y los graficamos
outlier_scores <- LOF(scale(data), k=5)

# Ordenamos los scores
outliers <- order(outlier_scores, decreasing=T)

# Juntamos los resultados
data_con_outliers <- cbind(OutlierScores=outlier_scores[outliers], data[outliers,])

# Identificamos los potenciales outliers (valores mayores a 1).
# Si tomamos estrictamente 1 como threshold, la mayoría de los puntos quedarían excluidos (>1700),
# por lo que necesitamos pensar en otro criterio para excluir los puntos
# Calculamos un threshold tomando como criterio de outlier los valores de LOF > Q3 + 3*IQR
q1 <- as.numeric(summary(outlier_scores)[2])
q3 <- as.numeric(summary(outlier_scores)[5])
iqr <- q3 - q1

upper_range = q3 + 3*iqr

potential_outliers <- data_con_outliers[data_con_outliers$OutlierScores > upper_range,]

cat("Hay", nrow(potential_outliers),"potenciales outliers utilizando como criterio los valores de LOF > Q3 + 3*IQR :\n")
potential_outliers

# dataset sin potenciales outliers
bcn_pisos_no_outliers <- bcn_pisos[data_con_outliers$OutlierScores < upper_range, ]
cat("\nDataset sin outliers (", nrow(bcn_pisos_no_outliers) ," observaciones)\n")
head(bcn_pisos_no_outliers)
```

#### 8. Efectúe un diagrama bivariante (plot) entre el “Valor de los pisos” y su “Superficie”. Calcule la covariancia y la correlación entre ambas variables.

```r
# Graficamos un diagrama bivariante
plot(
  bcn_pisos_no_outliers$Superf,
  bcn_pisos_no_outliers$Valor,
  xlab="Superficie",
  ylab="Valor",
  col="orange",
  main="Valor vs Superficie")

# Calculamos las medias para usarlas en el diagrama
mean_v <- mean(bcn_pisos_no_outliers$Valor)
mean_s <- mean(bcn_pisos_no_outliers$Superf)

# Agregamos líneas de medias
abline(v=mean_s, col="darkgreen", lw=2, lty=2:3)
text(x=mean_s + 120, y=1e+06, label=paste("Media (Valor): ", format(mean_v, nsmall=0)), col="darkgreen")

abline(h=mean_v, col="blue", lw=2, lty=2:3)
text(x=350, y=mean_v + 2e+06, label=paste("Media (Superficie): ", format(mean_s, nsmall=0)), col="blue")

# Hacemos lo mismo con la data estandarizada
plot(scale(bcn_pisos_no_outliers$Superf),
    scale(bcn_pisos_no_outliers$Valor),
    main="Relationship between prices and area (standardized prices)",
    xlab = "standardized area (z)",
    ylab = "Standardized prices (z)",
    pch=19,
    col="skyblue4")

abline(h =mean(scale(bcn_pisos_no_outliers$Valor)), col = "red", lw=2, lty=2:3)
abline(v = mean(scale(bcn_pisos_no_outliers$Superf)), col = "red", lw=2, lty=2:3)
```

```r
# Calculamos la covarianza
cov_val_sup <- cov(bcn_pisos_no_outliers$Valor, bcn_pisos_no_outliers$Superf)

cat("Valor de la covarianza entre Superficie & Valor: ", cov_val_sup)

# Calculamos la correlación

cor_val_sup <- cor(bcn_pisos_no_outliers$Valor, bcn_pisos_no_outliers$Superf)

cat("\nValor de la correlación entre Superficie & Valor: ", cor_val_sup)
```

#### 9. Calcule la asociación entre el “Valor de los pisos” y el “Distrito”.

```r
# En este caso queremos calcular la asociación entre una variable continua (Valor de los pisos) y una variable categórica (Distrito)
# Para ello realizamos el one-way analysis of mean

oneway.test(bcn_pisos_no_outliers$Valor~bcn_pisos_no_outliers$Dist,var.equal=T)
```

*Interpretación:* el alto valor de F y el valor de p-value extremamente bajo nos indican que podemos rechazar la hipótesis nula y concluir que hay una diferencia estadística significativa entre las medias (variable continua 'Valor') de los grupos de la variable categórica 'Distrito'.

#### 10. Calcule la asociación entre el “Distrito” y el “Estado de la vivienda”.

```r
# Calculamos la asociación entre dos variables categóricas: “Distrito” y el “Estado de la vivienda”,
# para ello realizamos el Pearson Chi-square test

chisq.test(bcn_pisos_no_outliers$Dist, bcn_pisos_no_outliers$Estat)
```

*Interpretación:* dado que el p-value es cercano al cero, podemos inferir que las variables son dependientes.
